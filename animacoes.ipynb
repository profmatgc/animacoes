# A1: **Nota√ß√µes de Derivada**

```python
%%manim -r 1080,1920 -qk -v WARNING --format mp4 notacoes_de_derivada

class notacoes_de_derivada(Scene): # Anima√ß√£o 1
    def construct(self):
        # Definindo as vari√°veis
        username = Text(
            '@profmatgc',
            font_size=65,
            fill_opacity=0.25
        ).move_to(6.5*DOWN)

        derivadas = [
            MathTex(
                "\\lim_{h \\to 0}",
                "\\frac{f(x+h)-f(x)}{h}"
            ).scale(2.5),
            MathTex("f‚Äô(x)").scale(5),
            MathTex("\dfrac{\mathrm{d}f(x)}{\mathrm{d}x}").scale(4),
            MathTex("\dot f(x)").scale(5),
            MathTex("D_x f(x)").scale(5)
        ]

        matematicos = ['Lagrange', 'Leibniz', 'Newton', 'Euler']

        textos = [
            Text(
                'Derivada pela \ndefini√ß√£o de Limite',
                font_size=92.5
            ).set_color_by_gradient(LIGHT_PINK, RED).move_to(5*UP)
        ]
        textos.extend([Text(nome, font_size=140).set_color_by_gradient(LIGHT_PINK, RED).move_to(5*UP) for nome in matematicos])

        # Cena inicial
        self.play(Write(derivadas[0]), Write(textos[0]), FadeIn(username), run_time=2)
        self.wait(2)

        # Apresentando cada nota√ß√£o de derivada e seu respectivo matem√°tico
        for i in range(1,5):
            self.play(Transform(derivadas[0], derivadas[i]), Transform(textos[0], textos[i]), run_time=1.75)
            self.wait()

        # Encerramento
        self.play(FadeOut(derivadas[0]), FadeOut(textos[0]), FadeOut(username), run_time=2)
```

# A2: **Soma de Riemman**

```python
%%manim -r 1080,1920 -qk -v WARNING --format mp4 soma_de_riemman

class soma_de_riemman(MovingCameraScene): # Anima√ß√£o 2
    def construct(self):
        # Definindo as vari√°veis
        titulo = Text(
            'Integral e a \nSoma de Riemman',
            font_size=100
        ).move_to(0.75*UP).set_color_by_gradient(LIGHT_PINK, RED)

        username = Text(
            '@profmatgc',
            font_size=65,
            fill_opacity=0.25
        ).move_to(6.5*DOWN)

        eixos = Axes(
            x_range=[0,11],
            x_length=10,
            y_range=[0,6],
            y_length=8
        ).add_coordinates()

        grafico = eixos.plot(
            lambda x: x**(1/2),
            x_range=[0,10],
            color=RED_D
        )

        funcao = MathTex(
            'f(x)=\sqrt{x}',
            color=RED_D
        ).move_to(0.55*UP+4.5*RIGHT).scale(0.5)

        medidas = [10, 5, 2, 1, 0.5, 0.1, 0.01]

        retangulos= VGroup(
            *[
                eixos.get_riemann_rectangles(
                    graph=grafico,
                    x_range=[0,10],
                    stroke_width=0.1,
                    stroke_color=WHITE,
                    dx=dx
                )
                for dx in medidas
            ]
        ).set_color_by_gradient(LIGHT_PINK, RED)

        area = retangulos[0]

        equacoes = [
            MathTex(
                r"\int f(x)\mathrm{d}x",
                r"=",
                r"\lim_{n \to \infty} \sum_{i=1}^{n}",
                r"\Delta x_i",
                r"f(x_i)",
                font_size=125
            ),
            MathTex(
                r"\text{√Årea}",
                r"\approx",
                r"\sum_{i=1}^{2} 5 \sqrt{x_i}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"\approx",
                r"\sum_{i=1}^{5} 2 \sqrt{x_i}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"\approx",
                r"\sum_{i=1}^{10} 1 \sqrt{x_i}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"\approx",
                r"\sum_{i=1}^{20} 0.5 \sqrt{x_i}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"\approx",
                r"\sum_{i=1}^{100} 0.1 \sqrt{x_i}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"\approx",
                r"\sum_{i=1}^{1000} 0.01 \sqrt{x_i}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\lim_{n \to \infty} \sum_{i=1}^{n} \Delta x_i \sqrt{x_i};",
                r"\Delta x_i\in \left[ 0,10 \right]",
                font_size=92.5
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\int_{0}^{10} \sqrt{x}\mathrm{d}x",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\int_{0}^{10} x^{\frac{1}{2}}\mathrm{d}x",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\frac{x^{\frac{1}{2}+1}}{\frac{1}{2}+1}\bigg|_{0}^{10}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\frac{x^{\frac{3}{2}}}{\frac{3}{2}}\bigg|_{0}^{10}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\frac{\sqrt{x^{3}}}{\frac{3}{2}}\bigg|_{0}^{10}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\frac{2}{3}\sqrt{x^{2}\cdot x}\bigg|_{0}^{10}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\frac{2}{3}\cdot x\sqrt{x}\bigg|_{0}^{10}",
                font_size=165
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\frac{2}{3}\cdot \left( 10\sqrt{10} - 0\sqrt{0}\right)",
                font_size=140
            ),
            MathTex(
                r"\text{√Årea}",
                r"=",
                r"\frac{2}{3}\cdot 10\sqrt{10}",
                font_size=165
            ),
            MathTex(
                r"\therefore \text{√Årea}",
                r"=",
                r"\frac{20\sqrt{10}}{3}",
                font_size=165
            )
        ]

        integral = [
            equacao.move_to(6*UP).scale(0.6).set_color_by_gradient(LIGHT_PINK, RED) for equacao in equacoes
        ]

        chave_integral = Brace(integral[0][0], UP)
        texto_integral = chave_integral.set_opacity(0.5).get_text('Integral')

        chave_riemman = Brace(integral[0][2:5], DOWN)
        texto_riemman = chave_riemman.set_opacity(0.5).get_text('Soma de Riemman')

        chave_soma = Brace(integral[0][2], UP)
        texto_soma = chave_soma.set_opacity(0.5).get_text(r'Soma das √°reas\\dos ret√¢ngulos')

        chave_base = Brace(integral[0][3], DOWN)
        texto_base = chave_base.set_opacity(0.5).get_text('Base')

        chave_altura = Brace(integral[0][4], DOWN)
        texto_altura = chave_altura.set_opacity(0.5).get_text('Altura')

        ct_integral = VGroup(
            chave_integral,
            texto_integral
        )

        ct_riemman = VGroup(
            chave_riemman,
            texto_riemman
        )

        ct_detalhes = VGroup(
            chave_soma,
            texto_soma,
            chave_base,
            texto_base,
            chave_altura,
            texto_altura
        )

        # Cena inicial
        self.play(Write(titulo), FadeIn(username), run_time=1.5)
        self.play(titulo.animate.shift(6*UP).scale(0.925), run_time=1.5)

        # Apresentando os principais elementos da anima√ß√£o
        self.play(
            ReplacementTransform(titulo, integral[0]),
            FadeIn(eixos),
            Create(grafico),
            Write(funcao),
            run_time=2
        )
        self.wait()

        # Salva a √∫ltima cena
        self.camera.frame.save_state()

        # Muda a posi√ß√£o da c√¢mera e apaga funcao
        self.play(
            self.camera.frame.animate.scale(0.4).move_to(6*UP+3.55*LEFT),
            FadeOut(funcao)
        )

        # Mostra o que √© a integral
        self.play(FadeIn(ct_integral), run_time=1.5)
        self.wait()

        # Muda a posi√ß√£o da c√¢mera e apaga ct_integral
        self.play(
            self.camera.frame.animate.scale(1.3).move_to(6*UP+2.15*RIGHT),
            FadeOut(ct_integral)
        )

        # Mostra o que √© a soma de riemman
        self.play(FadeIn(ct_riemman), run_time=1.5)
        self.wait()

        # Muda a posi√ß√£o da c√¢mera e transforma ct_riemman em ct_detalhes
        self.play(
            self.camera.frame.animate.scale(1.1).move_to(6*UP+2*RIGHT),
            ReplacementTransform(ct_riemman, ct_detalhes),
            run_time=1.5)
        self.wait(1.5)

        # Retorna na cena salva e apaga ct_detalhes
        self.play(
            Restore(self.camera.frame),
            FadeOut(ct_detalhes),
            run_time=1.5
        )
        self.wait(0.75)

        # Relacionando a √°rea abaixo da curva com a quantidade de ret√¢ngulos
        contador = 0
        for i in range(1, len(medidas)):
            nova_area = retangulos[i]
            self.play(
                Transform(area, nova_area),
                ReplacementTransform(integral[contador], integral[contador+1]),
                run_time=1.35
            )
            self.wait(0.5)
            contador += 1
        self.wait()

        # √änfase na soma de riemman e calcula da √°rea por meio da integral
        for j in range(contador, len(integral)-1):
            if j == 7:
                caixa = SurroundingRectangle(integral[j], buff=0.25, color=LIGHT_PINK)
                self.play(Create(caixa), run_time=1.5)
                self.wait()
                self.play(FadeOut(caixa), run_time=1.5)

            self.play(ReplacementTransform(integral[j], integral[j+1]), run_time=1.35)
            self.wait(0.5)

        # Encerramento
        self.wait(1.5)
        self.play(
            FadeOut(integral[j+1]),
            FadeOut(retangulos),
            FadeOut(eixos),
            FadeOut(grafico),
            FadeOut(username),
            run_time=4
        )
```
